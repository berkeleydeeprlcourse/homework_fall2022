```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eval_arr = np.loadtxt("../data/hw4_q3_obstacles_obstacles-cs285-v0_26-01-2023_21-49-25/eval.txt")
```

```python
plt.figure(figsize=(15, 5), dpi=80)
plt.suptitle('Obstacles, train_steps_per_iter 20, batch_size 1000, horizon 10')

plt.subplot(121)
plt.title("Evaluation Average Return")
plt.plot(eval_arr[1], label="Evaluation")
plt.xticks(list(range(len(eval_arr[0]))))
plt.xlabel("Iteration")
plt.ylabel("Eval Average Return")

plt.grid()
plt.legend()

plt.subplot(122)
plt.title("Iteration Duration")
plt.plot(eval_arr[3], label="Duration")
plt.xticks(list(range(len(eval_arr[0]))))
plt.xlabel("Iteration")
plt.ylabel("Iteration Duration")

plt.grid()
plt.legend()
```
