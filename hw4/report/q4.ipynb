```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eval_arr_hor5 = np.loadtxt("../data/hw4_q4_reacher_horizon5_reacher-cs285-v0_28-01-2023_09-31-54/eval.txt")
eval_arr_hor15 = np.loadtxt("../data/hw4_q4_reacher_horizon15_reacher-cs285-v0_28-01-2023_09-33-40/eval.txt")
eval_arr_hor30 = np.loadtxt("../data/hw4_q4_reacher_horizon30_reacher-cs285-v0_28-01-2023_10-06-21/eval.txt")
```

```python
plt.figure(figsize=(15, 5), dpi=80)
plt.suptitle('Reacher, train_steps_per_iter 1000, batch_size 800')

plt.subplot(121)
plt.title("Evaluation Average Return")
plt.plot(eval_arr_hor5[1], label="Horizon 5")
plt.fill_between(list(range(len(eval_arr_hor5[0]))), eval_arr_hor5[1]-eval_arr_hor5[3], eval_arr_hor5[1]+eval_arr_hor5[3],alpha=0.2)
plt.plot(eval_arr_hor15[1], label="Horizon 15")
plt.fill_between(list(range(len(eval_arr_hor15[0]))), eval_arr_hor15[1]-eval_arr_hor15[3], eval_arr_hor15[1]+eval_arr_hor15[3],alpha=0.2)
plt.plot(eval_arr_hor30[1], label="Horizon 30")
plt.fill_between(list(range(len(eval_arr_hor30[0]))), eval_arr_hor30[1]-eval_arr_hor30[3], eval_arr_hor30[1]+eval_arr_hor30[3],alpha=0.2)
plt.xticks(list(range(len(eval_arr_hor5[0]))))
plt.xlabel("Iteration")
plt.ylabel("Eval Average Return")

plt.grid()
plt.legend(loc='lower right')

plt.subplot(122)
plt.title("Iteration Duration")
plt.plot(eval_arr_hor5[3], label="Horizon 5")
plt.plot(eval_arr_hor15[3], label="Horizon 15")
plt.plot(eval_arr_hor30[3], label="Horizon 30")
plt.xticks(list(range(len(eval_arr_hor5[0]))))
plt.xlabel("Iteration")
plt.ylabel("Iteration Duration(seconds)")

plt.grid()
plt.legend()
```
