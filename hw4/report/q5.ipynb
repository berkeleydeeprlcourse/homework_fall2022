```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eval_arr = np.loadtxt("../data/hw4_q5_cheetah_random_cheetah-cs285-v0_31-01-2023_08-07-46/eval.txt")
```

```python
plt.figure(figsize=(15, 5), dpi=80)
plt.suptitle('Cheetah, Random, train_steps_per_iter 1500, batch_size 5000, horizon 15')

plt.subplot(121)
plt.title("Evaluation Average Return")
plt.plot(eval_arr[1], label="Evaluation")
plt.xticks(list(range(len(eval_arr[0]))))
plt.xlabel("Iteration")
plt.ylabel("Eval Average Return")

plt.grid()
plt.legend()

plt.subplot(122)
plt.title("Iteration Duration")
plt.plot(eval_arr[3], label="Duration")
plt.xticks(list(range(len(eval_arr[0]))))
plt.xlabel("Iteration")
plt.ylabel("Iteration Duration")

plt.grid()
plt.legend()
```
