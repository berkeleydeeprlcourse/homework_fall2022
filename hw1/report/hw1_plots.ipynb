```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
fig.suptitle("Behaviour Cloning - Walker2d")
fig.supxlabel("Network with 2 Hidden Layers of size 64, 10K training steps per iteration, LR=0.001")

x = np.array([0.0005, 0.0008, 0.0009, 0.001, 0.0012, 0.0014])
y = np.array([794, 985, 2331 ,4553, 3753, 5287])

ax1.set_xlabel('Learning Rate')
ax1.set_ylabel('Mean Rewards(Validation)')
ax1.set_title("LR vs Mean Rewards for 10.000 Steps")
ax1.set_xticks(x)
ax1.plot(x, y)

x = np.array([5000, 10000, 15000, 20000])
y = np.array([72, 4614, 4903, 5338])

ax2.set_xlabel('Training Steps')
ax2.set_ylabel('Mean Rewards(Validation)')
ax2.set_title("Training Steps vs Mean Rewards for LR=0.001")
ax2.set_xticks(x)
ax2.plot(x, y)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
fig.suptitle("DAgger - Walker2d")
fig.supxlabel("Network with 2 Hidden Layers of size 64, 2K training steps per iteration, LR=0.001")

x = np.array([0, 1, 2, 3, 4])
y = np.array([0.13, 0.05, 0.02, 0.015, 0.01])

ax1.set_xlabel('Iteration')
ax1.set_ylabel('MSE Loss')
ax1.set_title("Iteration vs MSE Loss")
ax1.set_xticks(x)
ax1.plot(x, y)

x = np.array([0, 1, 2, 3, 4])
y = np.array([138, 1016, 4904, 4701, 5370])
y_std=np.array([167, 733, 727, 1184, 47])

ax2.set_xlabel('Iteration')
ax2.set_ylabel('Mean Rewards')
ax1.set_title("Iteration vs Mean Rewards")
ax2.set_xticks(x)
ax2.errorbar(x, y, yerr=y_std, label="DAgger")
ax2.axhline(y = 5567, color = 'r', linestyle = '-',label="Expert")
ax2.axhline(y = 138, color = 'g', linestyle = '-',label="Behaviour Cloning")
ax2.legend()

plt.show()
```
