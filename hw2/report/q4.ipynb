```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
q4_b50K_r002_rtg_nn_baseline = np.loadtxt("q4/q4_b50K_r0.02_rtg_nn_baseline.txt")
```

```python
q4_b50K_r001_rtg_nn_baseline = np.loadtxt("q4/q4_b50K_r0.01_rtg_nn_baseline.txt")
```

```python
q4_b30K_r002_rtg_nn_baseline = np.loadtxt("q4/q4_b30K_r0.02_rtg_nn_baseline.txt")
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(20, 10))

fig.supxlabel("Params: --ep_len 150 --discount 0.95 -n 100 -l 2 -s 32 -rtg --nn_baseline")


x = np.array(range(len(q4_b50K_r002_rtg_nn_baseline)))

ax1.set_xlabel('Iteration')
ax1.set_ylabel('Mean Rewards(Validation)')
# ax1.set_title("Iteration vs Mean Rewards for 1000 Steps")
ax1.set_xticks(np.arange(0, len(x) + 1, 10))
ax1.axhline(y = 200.0, color = 'g', linestyle = '-')

ax1.plot(x, q4_b50K_r002_rtg_nn_baseline, label="q4_b50K_r002_rtg_nn_baseline", color="red")
ax1.plot(x, q4_b50K_r001_rtg_nn_baseline, label="q4_b50K_r001_rtg_nn_baseline", color="orange")

#ax1.plot(x, q4_b30K_r001_rtg_nn_baseline, label="q4_b30K_r001_rtg_nn_baseline", color="cyan")
#ax1.plot(x, q4_b30K_r002_nn_baseline, label="q4_b30K_r002_nn_baseline", color="magenta")
ax1.plot(x, q4_b30K_r002_rtg_nn_baseline, label="q4_b30K_r002_rtg_nn_baseline", color="blue")

plt.legend()
plt.show()
```

```python
q4_b50K_r002 = np.loadtxt("q4/q4_b50K_r0.02.txt")
q4_b50K_r002_nn_baseline = np.loadtxt("q4/q4_b50K_r0.02_nn_baseline.txt")
q4_b50K_r002_rtg = np.loadtxt("q4/q4_b50K_r0.02_rtg.txt")
q4_b50K_r002_rtg_nn_baseline = np.loadtxt("q4/q4_b50K_r0.02_rtg_nn_baseline.txt")
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(20, 10))

fig.supxlabel("Params: --ep_len 150 --discount 0.95 -n 100 -l 2 -s 32")


x = np.array(range(len(q4_b50K_r002)))

ax1.set_xlabel('Iteration')
ax1.set_ylabel('Mean Rewards(Validation)')
# ax1.set_title("Iteration vs Mean Rewards for 1000 Steps")
ax1.set_xticks(np.arange(0, len(x) + 1, 10))
ax1.axhline(y = 200.0, color = 'g', linestyle = '-')

ax1.plot(x, q4_b50K_r002, label="q4_b50K_r002", color="red")
ax1.plot(x, q4_b50K_r002_nn_baseline, label="q4_b50K_r002_nn_baseline", color="orange")
ax1.plot(x, q4_b50K_r002_rtg, label="q4_b50K_r002_rtg", color="blue")
ax1.plot(x, q4_b50K_r002_rtg_nn_baseline, label="q4_b50K_r002_rtg_nn_baseline", color="cyan")

plt.legend()
plt.show()
```
