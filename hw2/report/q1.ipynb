```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
q1_sb_no_rtg_dsa = np.loadtxt("q1/q1_sb_no_rtg_dsa.txt")
q1_sb_rtg_dsa = np.loadtxt("q1/q1_sb_rtg_dsa.txt")
q1_sb_rtg_na = np.loadtxt("q1/q1_sb_rtg_na.txt")
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(20, 10))

#fig.suptitle("PG - CartPole")
fig.supxlabel("Params: -lr 0.003 --seed 2 -l 2 --size 32")


x = np.array(range(len(q1_sb_no_rtg_dsa)))

ax1.set_xlabel('Iteration')
ax1.set_ylabel('Mean Rewards(Validation)')
# ax1.set_title("Iteration vs Mean Rewards for 1000 Steps")
ax1.set_xticks(np.arange(0, len(x) + 1, 10))

ax1.plot(x, q1_sb_no_rtg_dsa, label="q1_sb_no_rtg_dsa", color="red")
ax1.plot(x, q1_sb_rtg_dsa, label="q1_sb_rtg_dsa", color="black")
ax1.plot(x, q1_sb_rtg_na, label="q1_sb_rtg_na", color="green")

plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import ttest_ind

print("q1_sb_no_rtg_dsa mean", np.mean(q1_sb_no_rtg_dsa))
print("q1_sb_rtg_dsa mean", np.mean(q1_sb_rtg_dsa))
print("q1_sb_rtg_na mean", np.mean(q1_sb_rtg_na))

print("q1_sb_no_rtg_dsa <-> q1_sb_rtg_dsa : same mean if p_value > 0.05 ", ttest_ind(q1_sb_no_rtg_dsa, q1_sb_rtg_dsa).pvalue)
print("q1_sb_rtg_dsa <-> q1_sb_rtg_na : same mean if p_value > 0.05 ", ttest_ind(q1_sb_rtg_dsa, q1_sb_rtg_na).pvalue)
```
