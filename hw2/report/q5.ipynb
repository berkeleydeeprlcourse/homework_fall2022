```python
fig, ax1 = plt.subplots(1, 1, figsize=(20, 10))

fig.supxlabel("Params: --ep_len 1000 --discount 0.99 -n 300 -l 2 -s 32 -b 2000 -lr 0.001 --action_noise_std 0.5 --reward_to_go --nn_baseline")


x = np.array(range(len(q5_b2000_r0001_trg_nn_baseline)))

ax1.set_xlabel('Iteration')
ax1.set_ylabel('Mean Rewards(Validation)')
# ax1.set_title("Iteration vs Mean Rewards for 1000 Steps")
ax1.set_xticks(np.arange(0, len(x) + 1, 10))
ax1.axhline(y = 400.0, color = 'g', linestyle = '-')

#ax1.plot(x, q5_b2000_r0001_lambda0, label="q5_b2000_r0001_lambda0", color="magenta")

#ax1.plot(x, q5_b2000_r0001_lambda095, label="q5_b2000_r0001_lambda095", color="red")
#ax1.plot(x, q5_b2000_r0001_lambda098, label="q5_b2000_r0001_lambda098", color="orange")
#ax1.plot(x, q5_b2000_r0001_lambda099, label="q5_b2000_r0001_lambda099", color="blue")

ax1.plot(x, q5_b2000_r0001_lambda1, label="q5_b2000_r0001_lambda1", color="cyan")
ax1.plot(x, q5_b2000_r0001_trg_nn_baseline, label="q5_b2000_r0001_vanilla neural network baseline estimator (no noise)", color="purple")

plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(20, 10))

fig.supxlabel("Params: --ep_len 1000 --discount 0.99 -n 300 -l 2 -s 32 -b 2000 -lr 0.001 --action_noise_std 0.5 --reward_to_go --nn_baseline")


x = np.array(range(len(q5_b2000_r0001_trg_nn_baseline)))

ax1.set_xlabel('Iteration')
ax1.set_ylabel('Mean Rewards(Validation)')
# ax1.set_title("Iteration vs Mean Rewards for 1000 Steps")
ax1.set_xticks(np.arange(0, len(x) + 1, 10))
ax1.axhline(y = 400.0, color = 'g', linestyle = '-')

ax1.plot(x, q5_b2000_r0001_lambda0, label="q5_b2000_r0001_lambda0", color="red")
ax1.plot(x, q5_b2000_r0001_lambda095, label="q5_b2000_r0001_lambda095", color="magenta")
ax1.plot(x, q5_b2000_r0001_lambda098, label="q5_b2000_r0001_lambda098", color="orange")
ax1.plot(x, q5_b2000_r0001_lambda099, label="q5_b2000_r0001_lambda099", color="blue")

ax1.plot(x, q5_b2000_r0001_lambda1, label="q5_b2000_r0001_lambda1", color="cyan")

plt.legend()
plt.show()
```
