```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
q3_b40000_r0_005 = np.loadtxt("q3/q3_b40000_r0.005.txt")
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(20, 10))

#fig.suptitle("PG - CartPole")
fig.supxlabel("Params: -lr 0.005 -l 2 --size 64 -b 40.000")


x = np.array(range(len(q3_b40000_r0_005)))

ax1.set_xlabel('Iteration')
ax1.set_ylabel('Mean Rewards(Validation)')
# ax1.set_title("Iteration vs Mean Rewards for 1000 Steps")
ax1.set_xticks(np.arange(0, len(x) + 1, 10))
ax1.axhline(y = 180.0, color = 'g', linestyle = '-')

ax1.plot(x, q3_b40000_r0_005, label="q3_b40000_r0_005", color="black")

plt.legend()
plt.show()
```
